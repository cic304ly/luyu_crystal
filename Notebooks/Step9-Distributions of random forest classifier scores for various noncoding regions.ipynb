<h2>The distribution of predictive scores for variants was displayed within each functional category</h2>

<h3>This script corresponds to the "Classifier score across human non-coding regions" section in our manuscript.</h3>

```python
import re, os, sys
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
import pandas as pd
```

```python
datasets = pd.read_csv("../Classifier score across human non-coding regions/dataset_8terms_score.txt",sep = "\t")      
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(121)
sbn.violinplot(x="region", y="score", data=datasets)
plt.ylabel('Classifier score', fontdict={'family' : 'Times New Roman', 'size'   : 24})
plt.xlabel('Region', fontdict={'family' : 'Times New Roman', 'size'   : 24})
plt.yticks(fontproperties = 'Times New Roman', size = 22)
plt.xticks(fontproperties = 'Times New Roman', size = 22)
plt.xticks(rotation=30)
plt.axhline(0.256, 0, 100, color="blue", linestyle = 'dashed')
plt.show()
```
