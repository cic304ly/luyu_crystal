<h2> This script corresponds to the "Comparison of model for removing GWASdb and other methods" section of our manuscript. And this script runs from top to bottom, and the first part is finished before the second part is run.</h2>

```python
import re, os, sys
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import auc
```

```python
labellist = []
Ourpredictproba = []
f = open("../Other methods score2/validation_set_predict_proba",'rt')
for line in f:
	line = line.strip()
	li = re.split("\t",line)
	labellist.append(int(li[0]))
	Ourpredictproba.append(float(li[-1]))
f.close()
fpr, tpr, thresholds = metrics.roc_curve(labellist, Ourpredictproba, pos_label = 1)
AUC_our = metrics.auc(fpr, tpr)
```

<h4>Load CADD v1.4 and v1.6 prediction results</h4>

```python
##CADD v1.4
labellist_CADDv14 = []
CADDv14predictproba = []
f = open("../Other methods score2/validation_set_CADDv14__predict_proba",'rt')
for line in f:
	line = line.strip()
	li = re.split("\t",line)
	labellist_CADDv14.append(int(li[0]))
	CADDv14predictproba.append(float(li[-1]))
f.close()
fpr_CADDv14, tpr_CADDv14, thresholds_CADDv14 = metrics.roc_curve(labellist_CADDv14, CADDv14predictproba, pos_label = 1)
AUC_CADDv14 = metrics.auc(fpr_CADDv14, tpr_CADDv14)

##CADD v1.6
labellist_CADDv16 = []
CADDv16predictproba = []
f = open("../Other methods score2/validation_set_CADDv16__predict_proba",'rt')
for line in f:
	line = line.strip()
	li = re.split("\t",line)
	labellist_CADDv16.append(int(li[0]))
	CADDv16predictproba.append(float(li[-1]))
f.close()
fpr_CADDv16, tpr_CADDv16, thresholds_CADDv16 = metrics.roc_curve(labellist_CADDv16, CADDv16predictproba, pos_label = 1)
AUC_CADDv16 = metrics.auc(fpr_CADDv16, tpr_CADDv16)
```

<h4>Load GWAVA v1.0 prediction results</h4>

```python
##GWAVA
labellist_GWAVA = []
GWAVApredictproba_unmatched = []
GWAVApredictproba_TSS = []
GWAVApredictproba_region = []

f = open("../Other methods score2/validation_set_GWAVA_predict_proba",'rt')
for line in f:
	line = line.strip()
	li = re.split("\t",line)
	labellist_GWAVA.append(int(li[0]))
	GWAVApredictproba_unmatched.append(float(li[-1]))
	GWAVApredictproba_TSS.append(float(li[-2]))
	GWAVApredictproba_region.append(float(li[-3]))
f.close()
fpr_GWAVA_unmatched, tpr_GWAVA_unmatched, thresholds_GWAVA_unmatched = metrics.roc_curve(labellist_GWAVA, GWAVApredictproba_unmatched, pos_label = 1)
AUC_GWAVA_unmatched = metrics.auc(fpr_GWAVA_unmatched, tpr_GWAVA_unmatched)


fpr_GWAVA_TSS, tpr_GWAVA_TSS, thresholds_GWAVA_TSS = metrics.roc_curve(labellist_GWAVA, GWAVApredictproba_TSS, pos_label = 1)
AUC_GWAVA_TSS = metrics.auc(fpr_GWAVA_TSS, tpr_GWAVA_TSS)


fpr_GWAVA_region, tpr_GWAVA_region, thresholds_GWAVA_region = metrics.roc_curve(labellist_GWAVA, GWAVApredictproba_region, pos_label = 1)
AUC_GWAVA_region = metrics.auc(fpr_GWAVA_region, tpr_GWAVA_region)
```
