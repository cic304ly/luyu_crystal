<h2>Distribution of Gini importance of the 13 most important features in five training sets.</h2>

<h3>This script corresponds to the "Contribution of individual features" section of our manuscript.</h3>

```python
import seaborn as sbn
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv("../Contribution of individual features/13_impor_features.csv",sep = ",")   
```

```python
plt.figure(figsize=(8,8))
transform_Data = dataset.pivot(index='Features', columns='#Fivefold cross-validation',values='gini_importance').sort_index(axis = 0, ascending = False, by = [1])
sbn.heatmap(transform_Data,linewidths = 0.4, robust=True, annot=False, vmax=0.07, vmin=0, fmt='f', cmap='coolwarm')
cax = plt.gcf().axes[-1]
cax.tick_params(labelsize=24)
cax.set_label('colorbar')
plt.xlabel("Five-fold cross-validation",fontdict={'family' : 'Times New Roman', 'size'   : 32})
plt.ylabel("Features",fontdict={'family' : 'Times New Roman', 'size'   : 32})
plt.yticks(fontproperties = 'Times New Roman', size = 28)
plt.xticks(fontproperties = 'Times New Roman', size = 28)
plt.show()
```
